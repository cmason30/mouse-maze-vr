```
df_circular = pd.read_csv(r'circular.behavior', header=2,sep='\t')
df_corridor = pd.read_csv(r'corridor.behavior',header=2, sep='\t')
df_square = pd.read_csv(r'square.behavior', header=2, sep='\t')
df_ymaze = pd.read_csv(r'ymaze.behavior', header=2, sep='\t')
```

```
ymaze_coords = np.array([[-433.013, -452.582],
                         [0,-202.582],
                         [433.013, -452.582],
                         [558.013, -236.075],
                         [125, 13.925],
                         [125, 513.924],
                         [-125, 513.924],
                         [-125, 13.925],
                         [-558.013, -236.075]])


sq_coords = np.array([[750,-750],
                      [-750, -750],
                      [-750, 750],
                     [750, 750]])

circle = 750

corr_coords = np.array([[-75, -1332.286],
                        [-75, 1332.286],
                        [75, 1332.286],
                        [75, -1332.286]])
```

```
# Maze coordinates: input the type of maze you want. Maze types are given above or 
# accepts maze shaped in NumPy arrays given with [x,y] values
# !!!For circles it just needs the radius as an integer

# df: Give the file that has the mouse movement contained

# distance_threshold: set the threshold for unit distance the function considers the mouse to be close to the wall
# the minimum for this is 35 and considers the mouse touching the wall at that point. 

# Output: Outputs full DF with two added columns:
# First is df['mouse_dist'] which gives the distance of mouse from nearest wall
# Second is df['distance_marked'] which gives a 0 if mouse distance is above distance_threshold and a 1 if it is below distance_threshold

def mouse_edge_distance(maze_coordinates, df, distance_threshold=35):

  if isinstance(maze_coordinates, int):
    p = Point(0, 0)
    polygon_maze = p.buffer(maze_coordinates)

  else:
    polygon_maze = Polygon(maze_coordinates)

  df["mouse_dist"] = df.apply(lambda row: polygon_maze.exterior.distance(Point(row["Position.X"], row["Position.Y"])), axis=1)
  df['distance_marked'] = df.apply(lambda row: 1 if distance_threshold >= row['mouse_dist'] else 0, axis=1)

  return df

```

```
mouse_edge_distance(ymaze_coords,df_ymaze)
```

```
mouse_edge_distance(sq_coords, df_square)
```

```
mouse_edge_distance(circle, df_circular)
```

```
mouse_edge_distance(corr_coords, df_corridor)
```
